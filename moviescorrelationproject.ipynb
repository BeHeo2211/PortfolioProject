```python
#import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib 
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# read in the data
df = pd.read_csv(r'C:\Users\kienkauko\Documents\SQL Server Management Studio\db-beheo\movies.csv')

```

```python
# let look at the data
df.head()
```

```python
# we need to see if we have any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# data types for our colums
df.dtypes
```

```python
# change data type of colums
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Drop any duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# test correlation
# scatter plot with budget and company (cus i do believe that budget and company have high correlation w gross)

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('budget for firms')
plt.ylabel('Gross Earnings')
plt.show

```

```python
df
```

```python
# plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# lets start looking at correlation
df.corr(method='pearson') #pearson, kemdall, spearman
```

```python
# high correlation between budget and gross
# i was right
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.show()

plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Feature')
plt.ylabel('')
plt.show
```

```python
# Look at company
df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Feature')
plt.ylabel('Movie Feature')
plt.show
```
